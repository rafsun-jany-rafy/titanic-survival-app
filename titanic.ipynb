# ***Data Preprocessing***

As the cabin column has 687 missing values out of total 891 values, it is better to drop the column or feature

```python
df = df.drop(columns='Cabin')
```

```python
#first 5 rows
df.head()
```

```python
df.isnull().sum()
```

The Age column has 177 missing values, but the dataset has total 891 rows, so we cannot drop those rows. Again, age feature could be an important feature for the dataset, so it cannot be dropped. So, we can fill the missing values with mean

```python
df['Age'].mean()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

Now the Embarked column has two missing values, which we are going to fill by the mode

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].mode()[0]
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])#mode() returns a series, so [0] should be included
```

```python
df.isnull().sum()
```

Encoding the categorical data

```python
df = df.replace({'Sex': {'male':0, 'female':1}, 'Embarked': {'S':0,'C':1,'Q':2}})
```

```python
#df = df.replace({'Sex': {'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}})
```

```python
df.head()
```

# ***Splitting the dataset into features and target***

```python
X = df.drop(columns=['PassengerId','Name','Ticket','Survived'])
y = df['Survived']
```

# ***Model Selection and hyperparameter tuning using GridSearchCV***

```python
models_list = [LogisticRegression(max_iter=10000),SVC(), KNeighborsClassifier(),RandomForestClassifier()]
```

```python
hyperparameters = {
    
    'log_reg_parameters':{
        'C':[0.1,0.5,1,5,10,20],
        'solver': ['lbfgs', 'liblinear']
    },
    
    'svm_parameters':{
        'C': [0.1,0.5,1,5,10,20,100],
        'kernel': ['linear','poly','rbf','sigmoid'],
        'gamma': ['scale', 0.1, 0.01, 0.001]
    },
    
    'knn_parameters':{
        'n_neighbors': [3,5,7,9],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    
    'random_forest_parameters': {
        'n_estimators': [20,50,80,100,150,200],
        'criterion': ['gini','entropy','log_loss'],
        'max_depth': [3,4,5,6],
        'max_features': ['sqrt'],
        'min_samples_split':[5,10,15],
        'min_samples_leaf':[2,3,4,5,6],
    }
    
}
```

```python
hyperparameters.keys()
```

```python
hyperparameters.values()
```

```python
hyperparameters['knn_parameters']
```

```python
model_keys = list(hyperparameters.keys())
```

```python
print(model_keys)
```

```python
model_keys[0]
```

```python
def compare_models(list_of_models, dictionary_of_hyperparameters):
    
    results = []
    
    i = 0

    for model in list_of_models:
        
        key = model_keys[i]
        params = dictionary_of_hyperparameters[key]
        
        print(key)
        print(params)
        print('---------------------------------------------------')
        i = i+1
        
        classifier = GridSearchCV(model, params, cv=5)
        
        classifier.fit(X,y)
        
        results.append({
            'model_name': key,
            'best_score': classifier.best_score_,
            'best_hyperparameters': classifier.best_params_
        })
    
    pd.set_option('display.max_colwidth', None)  
    #results_df.iloc[-1]
    results_df = pd.DataFrame(results, columns=['model_name','best_score','best_hyperparameters'])
    return results_df
        
```

```python
compare_models(models_list, hyperparameters)
```

# The best score is found for RandomForestClassifier and best hyperparameters are criterion: 'entropy' and n_estimators: 100

```python
#Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=2)
```

```python
#loading the model
model = RandomForestClassifier(criterion= 'entropy', max_depth= 3, max_features= 'sqrt', min_samples_leaf= 5, min_samples_split= 10, n_estimators= 50, random_state=2)
```

```python
#fitting the model
model.fit(X_train,y_train)
```

# ***Model Evaluation***

```python
training_prediction = model.predict(X_train)
training_accuracy = metrics.accuracy_score(y_train, training_prediction)
print(f'Accuracy for training data: {training_accuracy}')
```

```python
testing_prediction = model.predict(X_test)
testing_accuracy = metrics.accuracy_score(y_test, testing_prediction)
print(f'Accuracy for the test data: {testing_accuracy}')
```

Add:

n_jobs=-1

classifier = GridSearchCV(
    model,
    params,
    cv=5,
    n_jobs=-1
)

This uses all CPU cores.

Reduce parameter space
Better Random Forest grid
'random_forest_parameters': {
    'n_estimators': [50,100],
    'criterion': ['gini','entropy'],
    'max_depth': [4,6],
    'min_samples_split':[5,10],
    'min_samples_leaf':[2,4]
}

Only:

2 * 2 * 2 * 2 * 2 = 32 combos

instead of 1080.

Better SVM grid
'svm_parameters':{
    'C': [0.1,1,10],
    'kernel': ['linear','rbf'],
    'gamma': ['scale',0.01]
}

Only:

3 * 2 * 2 = 12 combos

instead of 112.

Use RandomizedSearchCV instead

Much faster.

Example:

from sklearn.model_selection import RandomizedSearchCV

classifier = RandomizedSearchCV(
    model,
    params,
    cv=5,
    n_iter=20,
    n_jobs=-1,
    random_state=2
)

This tests only 20 random combinations.

Another issue

You are fitting on:

classifier.fit(X,y)

instead of train data.

Use:

classifier.fit(X_train, y_train)

Otherwise:

slower
data leakage
unrealistic scores
Recommended practical setup

For Titanic:

cv=3
n_jobs=-1
smaller grids
RandomizedSearchCV for RandomForest and SVM

That usually finishes in minutes instead of hours.
